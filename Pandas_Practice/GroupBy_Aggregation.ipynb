```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ali", "Rohit", "Neha", "Amit", "Sara", "John"],
    "Dept": ["IT", "HR", "IT", "Sales", "HR", "Sales"],
    "Salary": [50000, 40000, 55000, 60000, 42000, 58000],
    "Age": [21, 25, 20, 23, 22, 24]
})

print(df)
```

## 1Ô∏è‚É£ Basic GroupBy

### ‚úÖ Department wise average salary

```python
df.groupby("Dept")["Salary"].mean()
```

## 2Ô∏è‚É£ GroupBy with Sum
###‚úÖ Total salary per department

```python
df.groupby("Dept")["Salary"].sum()
```

## 3Ô∏è‚É£ GroupBy with Count
###‚úÖ ‡§ï‡§ø‡§§‡§®‡•á employees ‡§π‡•à‡§Ç ‡§π‡§∞ dept ‡§Æ‡•á‡§Ç?

```python
df.groupby("Dept")["Name"].count()
```

# 4Ô∏è‚É£ Multiple Aggregations (Real Analyst Style)

```python
df.groupby("Dept")["Salary"].agg(["mean","max","min","sum"])
```

## 5Ô∏è‚É£ GroupBy on Multiple Columns
###Dept + Age wise salary mean

```python
df.groupby(["Dept" , "Age"])['Salary'].mean()
```

## Department wise average age ‡§®‡§ø‡§ï‡§æ‡§≤‡•ã

```python
df.groupby("Dept")["Age"].mean()
```

# Department wise maximum salary ‡§®‡§ø‡§ï‡§æ‡§≤‡•ã

```python
df.groupby("Dept")["Salary"].max()
```

# ‡§π‡§∞ department ‡§Æ‡•á‡§Ç ‡§ï‡§ø‡§§‡§®‡•á ‡§≤‡•ã‡§ó ‡§π‡•à‡§Ç?

```python
df.groupby("Dept")["Name"].count()
```

# ‡§∏‡§¨‡§∏‡•á ‡§ú‡•ç‡§Ø‡§æ‡§¶‡§æ average salary ‡§ï‡§ø‡§∏ dept ‡§Æ‡•á‡§Ç ‡§π‡•à?

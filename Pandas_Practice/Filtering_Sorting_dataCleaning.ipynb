```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ali", "Rohit", "Neha", "Amit", "Sara"],
    "Age": [21, 25, 20, 23, 22],
    "Marks": [85, 90, 78, 88, 95],
    "City": ["Delhi", "Mumbai", "Pune", "Delhi", "Mumbai"]
})
print(df)
```

## Filtering

```python
# ✅ Marks > 85 वाले students निकालो

df[df["Marks"] > 85]
```

# ✅ City = Delhi वाले students

```python
df[df["City"] == "Delhi"]
```

# 2️⃣ Multiple Conditions Filtering (Real Analyst Work)

 ## Mumbai AND Marks > 90

```python
df[(df["City"] == "Mumbai" ) & (df["Marks"] > 90)]
```

## ✅ Age < 23 OR Marks > 90

```python
df[(df["Age"] < 23) | (df["Marks"] > 90)]
```

# 3️⃣ Sorting Data (Very Common)

## ✅ Marks के हिसाब से sort करो

```python
df.sort_values("Marks")
```

## ✅ Descending order

```python
df.sort_values("Marks" ,ascending=False)
```

# 4️⃣ Missing Values Handling (Data Cleaning)

```python
df2 = pd.DataFrame({
    "Name": ["Ali", "Rohit", "Neha"],
    "Marks": [85, None, 78]
})
df2
```

## ✅ Check null values

```python
df2.isnull().sum()
```

## ✅ Fill missing values

```python
df2["Marks"]=df2["Marks"].fillna(df2["Marks"].mean())
df2
```

## ✅ Drop missing rows

```python
df2.dropna()
```

# 5️⃣ Duplicate Data Remove


```python
df3 = pd.DataFrame({
    "Name": ["Ali", "Ali", "Rohit"],
    "Marks": [85, 85, 90]
})

df3.drop_duplicates()
```

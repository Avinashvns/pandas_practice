# loc and iloc Practice Set

```python
import pandas as pd

df=pd.DataFrame({
    "Name": ["Ali", "Rohit", "Neha", "Amit", "Sara"],
    "Age": [21, 25, 20, 23, 22],
    "Marks": [85, 90, 78, 88, 95],
    "City": ["Delhi", "Mumbai", "Pune", "Delhi", "Mumbai"]
})

df
```

```python
df.loc[0,'Marks']
```

```python
df.iloc[4,3]
```

```python
df.iloc[0]
```

```python
df.loc[4]
```

```python
# Row label 3 ‡§∏‡•á ‡§∏‡§ø‡§∞‡•ç‡§´ Name ‡§®‡§ø‡§ï‡§æ‡§≤‡•ã

df.loc[3,'Name']
```

```python
df.iloc[3,0]
```

# Row position 2 ‡§∏‡•á Marks ‡§®‡§ø‡§ï‡§æ‡§≤‡•ã


```python
df.iloc[2,2]
```

```python
df.loc[:,["Name","Marks"]]
```

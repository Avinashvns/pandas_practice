# Series

```python
import pandas as pd
marks=pd.Series([80,90,75,88])
print(marks)
```

# Column with Name Series

```python
import pandas as pd
marks=pd.Series([80,90,75,88], index=["A", "B" ,"C","D"])
print(marks)
```
